```python
def convert_min_to_float(x):
    try:
        minutes, seconds = map(int, x.split(':'))  
        return round(minutes + seconds / 60)  
    except (ValueError, AttributeError):
        return np.nan  # Handle invalid case (ex : NaN or wrong chain)

def convert_teams(x):
    if x in ['Golden St', 'e Warriors']: 
        return 'Golden State Warriors'
    elif x in ['Miami H', 'Miami He']: 
        return 'Miami Heat'
    else: 
        return x
    
def clean_data(df):
    self.check_nan()
    df['FG'] = pd.to_numeric(df['FG'],errors='coerce')
    df['MP'] = df['MP'].apply(convert_min_to_float)
    df['Team'] = df['Team'].apply(convert_teams)
    df['Against'] = df['Against'].apply(convert_teams)
    #self.df.save_csv(self.datapath+'data_cleaned.parquet')
    return df
```

```python
df_cleaned = clean_data(season21)
```
